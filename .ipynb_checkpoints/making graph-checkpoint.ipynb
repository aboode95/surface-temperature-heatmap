# Moving Average and Median filter

```python
filter_length = 51
t2_moving_average = np.convolve(df['t2'], np.ones((filter_length)), mode='same')
t2_moving_average /= filter_length
plt.plot(df['t2'])
plt.plot(t2_moving_average)
```

```python
from scipy.signal import medfilt
```

```python
t2_medfilt = medfilt(df['t2'], filter_length)
```

```python
plt.plot(df['t2'],  label="t2")
# plt.plot(t2_moving_average)
plt.plot(t2_medfilt, label='median filter')

plt.title('median filter on t2')
plt.legend()
plt.show()
```

# Graphing t2 from Average of 5 data points

```python
df
```

```python
# Average of first 5 data points
average = sum(df['t2'].iloc[0:5])/5
print(average)
```

```python
# Printing time after the 5 data points
time = df['time'].iloc[4]
print(time)
```

```python
# Looping through the data set and averaging consecutive 5 data points
sample_size = 20
t2_average_array = []
f = 0
i = 0
while f < 3040/sample_size:
    average = sum(df['t2'].iloc[i:i+sample_size])/sample_size
    t2_average_array = np.append(t2_average_array, average)
    i += sample_size
    f+=1
```

```python
plt.plot(df['t2'])
plt.plot(range(0,3040,sample_size), t2_average_array)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, figsize=(6,4))
fig.suptitle('Different Data filtration')

ax1.plot(df['t2'], label='raw data')
ax1.set_title('Raw Data', loc='right')
ax2.plot(t2_moving_average, 'tab:orange')
ax2.set_title('Moving Average', loc='right')
ax3.plot(t2_medfilt, 'tab:green')
ax3.set_title('Median Filter', loc='right')
ax4.plot(range(0,3040,sample_size), t2_average_array, 'tab:red')
ax4.set_title('Manual Average', loc='right')

# Hiding x labels and tick labels for first 3 plots  
ax1.label_outer()
ax2.label_outer()
ax3.label_outer()

plt.subplots_adjust(hspace=0.5)
plt.show()
```

### [Code Explanation Link](https://matplotlib.org/devdocs/gallery/subplots_axes_and_figures/subplots_demo.html)

```python
def manual_average(mylist = []):
    sample_size = 20
    average_array = []
    f = 0
    i = 0
    while f < 3040/sample_size:
        average = sum(mylist.iloc[i:i+sample_size])/sample_size
        average_array = np.append(average_array, average)
        i += sample_size
        f+=1
    return average_array
```

```python
fig, axs = plt.subplots(3, 2, figsize=(20,5))

axs[0, 0].plot(manual_average(df['t0']))
axs[0, 0].set_title('t0')

axs[0, 1].plot(manual_average(df['t1']), 'tab:orange')
axs[0, 1].set_title('t1')

axs[1, 0].plot(manual_average(df['t2']), 'tab:green')
axs[1, 0].set_title('t2')

axs[1, 1].plot(manual_average(df['t3']), 'tab:red')
axs[1, 1].set_title('t3')

axs[2, 0].plot(manual_average(df['t4']), 'tab:cyan')
axs[2, 0].set_title('t4')

for ax in axs.flat:
    ax.set(xlabel='time-series', ylabel='temperature')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
fig.suptitle('Temperature of all Points', fontsize=20)
plt.subplots_adjust(wspace=0.1, hspace=0.5)
plt.show()
```

```python
plt.plot(manual_average(df['t0']), label="Leg Right")
plt.plot(manual_average(df['t1']), label="Leg Left")
plt.plot(manual_average(df['t2']), label="MIDPOINT")
plt.plot(manual_average(df['t3']), label="Head Right")
plt.plot(manual_average(df['t4']), label="Head Left")

plt.legend()
```

## Ecludian Distance Transform Matrix

### Inputing sensor Location

```python
# making a zero matrix
sen = np.zeros([5,5])
print(sen)
```

```python
# placing sen at position [2,2]
sen[2,2] = 1
print(sen)
```

```python
# Euclidian distance transforming the sensor array to find the relative position of the other points from the sensor.
from scipy import ndimage
np.set_printoptions(precision=3)  # Easier to read the result with fewer digits.

sen_euc = ndimage.distance_transform_edt(np.logical_not(sen))

print(sen_euc)
```

```python
# Plotting the array with the euclidian distance of all points from the sensor

fig, axs = plt.subplots(1,5, figsize=(15,3))

axs[0].imshow(sen_euc, cmap='Greys_r') # Display data as an image, i.e., on a 2D regular raster.
axs[0].set_title('none')
axs[1].imshow(sen_euc, cmap='Greys_r', interpolation='sinc')
axs[1].set_title('sinc')
axs[2].imshow(sen_euc, cmap='Greys_r', interpolation='gaussian')
axs[2].set_title('gaussian')
axs[3].imshow(sen_euc, cmap='Greys_r', interpolation='spline36')
axs[3].set_title('spline36')
axs[4].imshow(sen_euc, cmap='Greys_r', interpolation='bicubic')
axs[4].set_title('bicubic')

fig.suptitle('interpolation of imshow()')
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
# plt.imshow(sen_euc, cmap='magma')
```

```python
im = plt.imshow(sen_euc, cmap='Greys_r')
plt.colorbar(im)
```

### Some random example for plotting

```python
# Example from: https://stackoverflow.com/questions/6238250/multivariate-spline-interpolation-in-python-scipy

data = np.arange(40).reshape((8,5)).astype(float)
print(data)

coords = np.array([[1, 3], [6, 2], [7, 3], [3, 3]])  # [y,x] coordinates. Why?
print(coords)
coords = coords.T
print(coords)

row, column = coords
print(row)
print(column)

zi = ndimage.map_coordinates(data, coords, order=3, mode='nearest')

nrows, ncols = data.shape
print(nrows)
print(ncols)
im = plt.imshow(data, interpolation='nearest', extent=[0, ncols, nrows, 0])
plt.colorbar(im)

plt.scatter(column, row, vmin=data.min(), vmax=data.max())
for r, c, z in zip(row, column, zi):
    plt.annotate('%0.3f' % z, (c,r), xytext=(-10,10), textcoords='offset points',
            arrowprops=dict(arrowstyle='->'), ha='right')
plt.show()
```
